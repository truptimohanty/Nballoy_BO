## Two dimensional UMAP representations of the alloy compostitons (Experimental as well as Suggested)


```python
import pandas as pd
import numpy as np
```

### UMAP representation for UTS

```python
df_all = pd.read_csv('data/EI_UMaP/Nb_alloy_UTS_roomtemp.csv')
df_all.head()
```

```python
df_all_features = df_all[['formula','temp','UTS_Mpa']]
df_all_features.columns = ['formula','T','target']
df_all_features
```

### Featurization 

```python
from cbfv.composition import generate_features
X_feature_all, y_all, formulae_exp, skipped_exp= generate_features(df_all_features,elem_prop='f3_revised',drop_duplicates=False,extend_features=False)

X_feature_all
```

```python
#print(X_train_unscaled.columns)
from sklearn.preprocessing import StandardScaler
total_columns = int((X_feature_all.shape[1])/3) 
X_featue_all_avg = X_feature_all.iloc[:,0:total_columns]



scaler = StandardScaler()
X_featue_all_avg_scaled = scaler.fit_transform(X_featue_all_avg)
X_featue_all_avg
```

```python
import umap
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

reducer = umap.UMAP(random_state=56, n_neighbors = 7)
ar_umap=reducer.fit_transform(X_featue_all_avg_scaled)
ar_umap
plt.figure(figsize=(8,8))
plt.scatter(ar_umap[0:17:,0],ar_umap[:17,1],label='Experimental Data', s=df_all['UTS_Mpa'][:17]/2, alpha=.7)
plt.scatter(ar_umap[17:24,0],ar_umap[17:24,1],label='Candidate Selection',s=df_all['UTS_Mpa'][17:24]/2,alpha=.7)
plt.legend(prop={'size': 16})
plt.xlabel('UMAP 1',size=14)
plt.ylabel('UMAP 2',size=14)

plt.title("UMAP by features for UTS",size=14)
plt.annotate('Nb87.7W11.2Zr1.1',(6.2, 10.65),c='k',fontsize=14)

ellipse = Ellipse((7.33,10.55),width=0.33,
        height=0.19,angle = 0,fill=False,edgecolor='red',lw=2, linestyle = '--')

plt.gca().add_patch(ellipse)


plt.annotate('Nb87.0Zr1.1Hf0.6W11.3',(5.6, 11.05),c='k',fontsize=14)

# circle = plt.Circle((6.92, 10.8),0.1,facecolor='None', edgecolor='red', linestyle = '--')
ellipse = Ellipse((6.95,10.96),width=0.3,
        height=0.16,angle = 0,fill=False,edgecolor='red',lw=2, linestyle = '--')

plt.gca().add_patch(ellipse)


plt.tick_params(labelsize=14)
plt.show()
df_all_features.iloc[17]

ar_umap[17]     
```
