```python
import pandas as pd
import numpy as np
```

### UMAP representation for UTS

```python
df_all = pd.read_csv('data/EI_UMaP/Nb_alloy_UTS_roomtemp.csv')
df_all.head()
```

```python
df_all['r']= df_all.groupby("Exp_the", as_index=False)["UTS_Mpa"].rank(ascending=False)
df_all.head()
```

```python
df_all_features = df_all[['formula','temp','UTS_Mpa']]
df_all_features.columns = ['formula','T','target']
df_all_features.head()
```

```python
## added now
df_uts = pd.read_csv('data/EI_UMAP/train_ultimatestrength_all.csv')
df_uts.max()
df_uts.head()
```

### Featurization 

```python
from cbfv.composition import generate_features
X_feature_all, y_all, formulae_exp, skipped_exp= generate_features(df_all_features,elem_prop='f3_revised',drop_duplicates=False,extend_features=False)

X_feature_uts,y_uts,formulae_uts, skipped_uts = generate_features(df_uts,elem_prop='f3_revised',drop_duplicates=False,extend_features=False)
X_feature_uts
```

```python
#print(X_train_unscaled.columns)
from sklearn.preprocessing import StandardScaler
total_columns = int((X_feature_all.shape[1])/3) 

X_featue_all_avg = X_feature_all.iloc[:,0:total_columns]

X_feature_uts_avg = X_feature_uts.iloc[:,0:total_columns]

scaler = StandardScaler()
scaler.fit(X_feature_uts_avg)
X_featue_all_avg_scaled = scaler.transform(X_featue_all_avg)
```

```python
from scipy.spatial.distance import cosine

highest = X_featue_all_avg_scaled[15]

distance_cosine = [cosine(X_featue_all_avg.iloc[15,:],X_featue_all_avg.iloc[i,:]) for i in range(0,24)]


def calculate_dist(a,b):
    return (np.linalg.norm(a-b))

distance_euc= [calculate_dist(X_featue_all_avg_scaled[15],X_featue_all_avg_scaled[i]) for i in range(0,24)]



distance_euc_act= [calculate_dist(X_featue_all_avg.iloc[15,:],X_featue_all_avg.iloc[i,:]) for i in range(0,24)]
print('EEEEEEEEEEEEEE ACTUAL')
print(distance_euc_act)

print('EEEEEEEEEEEEEE SCALED')
print(distance_euc)

```
