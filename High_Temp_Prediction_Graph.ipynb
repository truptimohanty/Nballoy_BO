```python

import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np    



## Read the original dataset
df_UTS = pd.read_csv('data/HT_predictions/Nb_alloy_UTS_atmpct_suggested_quaternary.csv')
# df_UTS = df_UTS.sort_values('target')

plt.figure(figsize=(12,10))



colors = ['#cccccc', '#f77189', '#ce9032', '#97a431', '#66a182', '#cc7722', '#d1495b', '#00798c']
# Set your custom color palette
customPalette = sns.set_palette(sns.color_palette(colors))



gfg = sns.scatterplot(data = df_UTS, x = "T", y = "target" , 
                hue = "Expt", style="Expt", s = 200, alpha= 1, palette=customPalette, legend='auto')






plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), 
           markerscale = 1.5, 
          fontsize = 15)

plt.xlabel('Temperature ($^oC$)',fontsize=20)
plt.ylabel('Ultimate Tensile Strength (MPa)',fontsize=20)
plt.xlim((0,1900))
plt.ylim((0,1200))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.annotate('Nb87.7W11.2Zr1.1',(45, 1170),c='k',fontsize=14)

plt.show()

plt.figure(figsize=(1.5,4))

df_UTS['T_jitter'] = df_UTS['T'] #+ np.random.normal(0, 0.4, len(df_UTS['T']))

gfg = sns.scatterplot(data = df_UTS, x = "T_jitter", y = "target" , 
                hue = "Expt", style="Expt", s = 300, alpha= 1)



plt.legend([],[], frameon=False)

plt.xlabel('Temperature ($^oC$)',fontsize=20)
plt.ylabel('UTS (MPa)',fontsize=20)
plt.xlim((1199,1201))
plt.ylim((436,452))
# plt.yscale("log")
plt.xticks(fontsize=20, labels = ['',1200,''], ticks = [1199, 1200, 1201])
plt.yticks(fontsize=20)




plt.xticks(rotation=45)

plt.show()
```
