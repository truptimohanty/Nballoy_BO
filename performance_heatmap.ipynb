## Performance Evaluation for Different ML models 

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('data/model_output/r2score.csv')
#df.set_index('Models')
df = df.round(2)
df
```

```python
df.pivot("ML models",'Features','r_score_uts')
```

```python
df_heatmap = df.pivot("ML models",'Features','r_score_uts')
df_heatmap[["Composition","Oliynyk","Custom"]]
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

df_heatmap = df.pivot("ML models",'Features','r_score_uts')[["Composition","Oliynyk","Custom"]]

r_uts = sns.heatmap(df_heatmap,annot=True,cmap="flare",fmt='g',annot_kws={"size": 16},
                    linewidths=0.6, ax = axs[0,0])

r_uts.set_xticklabels(["Atom%","Oliynyk","Custom"], fontsize = 14)
r_uts.set_yticklabels(r_uts.get_ymajorticklabels(), fontsize = 14)
r_uts.set(xlabel=None, ylabel=None)
r_uts.set_title(r"$R^2$ Score UTS", fontsize = 14)


df_heatmap = df.pivot("ML models",'Features','MAE_uts')[["Composition","Oliynyk","Custom"]]
m_uts = sns.heatmap(df_heatmap,annot=True,cmap="flare_r",fmt='g',annot_kws={"size": 16},
                    linewidths=0.6, ax = axs[0,1])
m_uts.set_xticklabels(["Atom%","Oliynyk","Custom"], fontsize = 14)
m_uts.set_yticklabels(m_uts.get_ymajorticklabels(), fontsize = 14)
m_uts.set(xlabel=None, ylabel=None)
m_uts.set_title("MAE Score UTS", fontsize = 14)


df_heatmap = df.pivot("ML models",'Features','r_score_ys')[["Composition","Oliynyk","Custom"]]

r_ys = sns.heatmap(df_heatmap,annot=True,cmap="flare",fmt='g',annot_kws={"size": 16},
                   linewidths=0.6, ax = axs[1,0])
r_ys.set_xticklabels(["Atom%","Oliynyk","Custom"], fontsize = 14)
r_ys.set_yticklabels(r_ys.get_ymajorticklabels(), fontsize = 14)
r_ys.set(xlabel=None, ylabel=None)
r_ys.set_title("$R^2$ Score YS", fontsize = 14)


df_heatmap = df.pivot("ML models",'Features','MAE_ys')[["Composition","Oliynyk","Custom"]]

m_ys = sns.heatmap(df_heatmap,annot=True,cmap="flare_r",fmt='g',annot_kws={"size": 16},
                   linewidths=0.6, ax = axs[1, 1])
m_ys.set_xticklabels(["Atom%","Oliynyk","Custom"], fontsize = 14)
m_ys.set_yticklabels(m_ys.get_ymajorticklabels(), fontsize = 14)
m_ys.set(xlabel=None, ylabel=None)
m_ys.set_title("MAE Score YS", fontsize = 14)
```
