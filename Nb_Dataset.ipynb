## VISUALISATION OF Nb Alloy Experimental Data 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np       
```

```python
## Read the original dataset
df_UTS = pd.read_csv('data/original_data/Nb_alloy_UTS_atmpct.csv')

plt.rcParams['axes.linewidth'] =1.2

plt.figure(figsize=(6,6))
df_UTS['target'].hist(bins =30,alpha=0.9)
plt.ylabel('Counts',fontsize=15)
plt.xlabel('UTS (MPa)',fontsize=15)

plt.figure(figsize=(12,10))

gfg = sns.scatterplot(data = df_UTS, x = "T", y = "target" , 
                hue = "formula", style="formula", s = 200, alpha= 1)


plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), 
           markerscale = 1.5, edgecolor = "#FFFFFF",
          fontsize = 15)

plt.xlabel('Temperature ($^oC$)',fontsize=20)
plt.ylabel('Ultimate Tensile Strength (MPa)',fontsize=20)

plt.xlim((0,1900))
plt.ylim((0,1200))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.annotate('Nb87.7W11.2Zr1.1',(45, 1170),c='k',fontsize=14)
plt.annotate('Nb76.2Mo5Ti10W7.8Zr1',(45, 860),c='k',fontsize=14)
plt.annotate('Nb85.5Mo5.2W8.2Zr1.1',(45, 810),c='k',fontsize=14)

plt.savefig('figure/UTS.png', dpi=300, format='png', bbox_inches='tight')





## Read the original dataset
df_YS = pd.read_csv('data/original_data/Nb_alloy_YS_atmpct.csv')


plt.figure(figsize=(6,6))
df_YS['target'].hist(bins =30,alpha=0.9)
plt.ylabel('Counts',fontsize=15)
plt.xlabel('Yield Strength (MPa)',fontsize=15)




plt.figure(figsize=(12,10))

gfg = sns.scatterplot(data = df_YS, x = "T", y = "target" , 
                hue = "formula", style="formula", s = 200, alpha= 1)


plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), 
           markerscale = 1.5, edgecolor = "#FFFFFF",
          fontsize = 15)

plt.xlabel('Temperature ($^oC$)',fontsize=20)
plt.ylabel('Yield Strength (MPa)',fontsize=20)
plt.xlim((0,1900))
plt.ylim((0,1000))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)


plt.annotate('Nb87.7W11.2Zr1.1',(45, 970),c='k',fontsize=14)
plt.annotate('Nb85.5Mo5.2W8.2Zr1.1',(45, 690),c='k',fontsize=14)

plt.savefig('figure/YS.png', dpi=300, format='png', bbox_inches='tight')

plt.show()


```
